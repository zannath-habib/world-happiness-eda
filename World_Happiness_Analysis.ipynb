```python
#2. Importing lodged libraries in python for statistical analysis.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#3. Printing the dataset in python.
df=pd.read_csv("World Happiness 2019.csv")
df.head(15)
```

```python
#4. Bar Graph on Happiness
df.head(10).plot(x="Country", y="Happiness Score", kind="bar", figsize=(10,5), color="black", edgecolor="orange", lw=2.5)
plt.title("Top 10 Happy Countries")
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.show()
```

```python
#sorting on basis of Healthy Life
df1=df.sort_values(by="Healthy life", ascending=False)
```

```python
#5. Bar Graph on Healthy Life
df1.tail(10).plot(x="Country", y="Healthy life", kind="bar", figsize=(10,5), color="black", edgecolor="green", lw=2.5)
plt.title("Low Health Countries")
plt.xlabel("Country")
plt.ylabel("Health Score")
plt.show()
```

```python


df.sort_values(by="Healthy life", ascending=False)
```

```python
#6. Scatter Plot on GDP and Happiness Score
plt.scatter(df["Happiness Score"], df["GDP per capita"], marker="*", s=50, color="pink")
plt.title("Relation between GDP per capita vs Happiness Score")
plt.xlabel("Happiness Score")
plt.ylabel("GDP per capita")
plt.show()
```

```python
#7. Pie Chart of Top 15 Country based on GDP per capita
df_gdp = df.sort_values(by="Corruption", ascending=False).head(15)
plt.figure(figsize=(10, 5))
plt.pie(df_gdp["Corruption"], labels=df_gdp["Country"], autopct="%1.1f%%")
plt.title("Top 15 Countries by Corruption")
plt.show()
```

```python
plt.scatter(df["Happiness Score"], df["Freedom"], marker="*", s=50, color="pink")
plt.title("Relation between Happiness Score vs Freedom")
plt.xlabel("Happiness Score")
plt.ylabel("Freedom")
plt.show()
```

```python
#Least Corrupted Countries
df_corruption = df.sort_values(by="Corruption", ascending=False).head(15)
plt.plot(df_corruption["Country"], df_corruption["Corruption"], marker="o", linestyle="-", color="navy")
plt.title("Least Corrupted Countries")
plt.xlabel("Country")
plt.ylabel("Corruption")
plt.xticks(rotation=90)
plt.show()
```
